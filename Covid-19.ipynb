```python
import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv')
df.head()
```

```python
print(df.loc[df['Date'] == '04/03/20']) # 31
```

```python
df = df.drop(columns=['State/UnionTerritory','Sno'])
```

```python
df.head()
```

```python
df1 = df.groupby('Date',sort=False).sum()
```

```python
df1.head()
```

```python
df1['TotalCases'] = df1.sum(axis=1).astype('int')
```

```python
df1.head()
```

```python
df1
```

```python
plt.figure(figsize=(20,20))
plt.plot(df1.TotalCases , color='blue')
plt.xticks(rotation=90 , fontsize=15)
plt.show()
```

```python
df1.tail(30)
```

```python
rate = []
for i in range(df1.shape[0]-1):
     rate1 = (df1['TotalCases'].iloc[i+1]  -  df1['TotalCases'].iloc[i]) /  df1['TotalCases'].iloc[i]
     rate.append(rate1)
```

```python
df1['TotalCases'].iloc[51]
```

```python
avg = np.average(rate)
avg
```
